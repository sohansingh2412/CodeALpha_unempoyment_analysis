```python
import pandas as pd
import plotly.express as px

# 1. File read karein
df = pd.read_csv('Unemployment in India.csv')

# 2. Data cleaning (Columns ke extra spaces hatane ke liye)
df.columns = df.columns.str.strip()

# 3. Data ko group karein (Yeh line miss hone se error aa raha tha)
region_stats = df.groupby("Region")[["Estimated Unemployment Rate (%)"]].mean().reset_index()

# 4. Chart banana
fig = px.bar(region_stats, x='Region', y='Estimated Unemployment Rate (%)', 
             title="Average Unemployment Rate by Region", 
             color='Estimated Unemployment Rate (%)',
             color_continuous_scale='Reds')

fig.show()
```
